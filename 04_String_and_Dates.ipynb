```python
# Emp Data & Schema

emp_data = [
    ["001","101","John Doe","30","Male","50000","2015-01-01"],
    ["002","101","Jane Smith","25","Female","45000","2016-02-15"],
    ["003","102","Bob Brown","35","Male","55000","2014-05-01"],
    ["004","102","Alice Lee","28","Female","48000","2017-09-30"],
    ["005","103","Jack Chan","40","Male","60000","2013-04-01"],
    ["006","103","Jill Wong","32","Female","52000","2018-07-01"],
    ["007","101","James Johnson","42","Male","70000","2012-03-15"],
    ["008","102","Kate Kim","29","Female","51000","2019-10-01"],
    ["009","103","Tom Tan","33","Male","58000","2016-06-01"],
    ["010","104","Lisa Lee","27","Female","47000","2018-08-01"],
    ["011","104","David Park","38","Male","65000","2015-11-01"],
    ["012","105","Susan Chen","31","Female","54000","2017-02-15"],
    ["013","106","Brian Kim","45","Male","75000","2011-07-01"],
    ["014","107","Emily Lee","26","Female","46000","2019-01-01"],
    ["015","106","Michael Lee","37","Male","63000","2014-09-30"],
    ["016","107","Kelly Zhang","30","Female","49000","2018-04-01"],
    ["017","105","George Wang","34","Male","57000","2016-03-15"],
    ["018","104","Nancy Liu","29","","50000","2017-06-01"],
    ["019","103","Steven Chen","36","Male","62000","2015-08-01"],
    ["020","102","Grace Kim","32","Female","53000","2018-11-01"]
]

emp_schema = "employee_id string, department_id string, name string, age string, gender string, salary string, hire_date string"
```

```python
#create dataframe
emp = spark.createDataFrame(data=emp_data, schema=emp_schema)
```

```python
#show dataframe(Action)
emp.show()
```

```python
emp.printSchema()
```

```python
#case when
from pyspark.sql.functions import when, col
emp_gender_fixed=emp.withColumn("new gender",when(col("gender") == "Male", "M")
                                .when(col("gender")=="Female", "F")
                                .otherwise(None)
                               )
```

```python
emp_gender_fixed.show()
```

```python
from pyspark.sql.functions import when,col
age_category=emp.withColumn("age_group", when(col("age")<30,"Young")
                            .when((col("age")>=30) & (col("age")<=50),"Adult")
                           
                            .otherwise("senior")
                           )
```

```python
age_category.show()
```

```python
from pyspark.sql.functions import to_date, col
emp = emp.withColumn("hire_date",to_date(col("hire_date"),"yyyy-MM-dd")
                    )
```

```python
emp.printSchema()
```

```python
from pyspark.sql.functions import expr
emp=emp.withColumn("salary_category",expr("case when salary >=70000 then 'High' when salary >=50000 then 'Medium' else 'low' END")
                  )
```

```python
emp.show()
```

```python
#replace in strings
from pyspark.sql.functions import regexp_replace
emp_name_fixed = emp.withColumn("new_name",regexp_replace(col("name"),"J","z"))
```

```python
emp_name_fixed.show()
```

```python
#Add date columns
from pyspark.sql.functions import current_date, current_timestamp
emp_dated=emp.withColumn("DATE_NOW", current_date()).withColumn("timesamp_now",current_timestamp())
```

```python
emp_dated.show(truncate=False)
```

```python
#drop null gender records
emp_1=emp_dated.na.drop()
```

```python
emp_1.show()
```

```python
#fix null values
from pyspark.sql.functions import coalesce, lit

emp_null_df=emp_gender_fixed.withColumn("new gender",coalesce(col("new gender"), lit("o")))
```

```python
emp_null_df.show()
```

```python
emp_final=emp_name_fixed.drop("name").withColumnRenamed("new_name","name")
```

```python
emp_final.show()
```
